# Adult Income Dataset: What Influences Earning Over $50K?

This analysis explores the Adult Income dataset to uncover what factors are most associated with earning more than $50,000 per year. We'll focus on education, gender, age, and work hours.

## 1. Imports & Setup

Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load Dataset

Load the CSV file and take a first look at the structure and contents.

```python
df = pd.read_csv("adult.csv")  
df.head()
```

## 3. Initial Exploration & Cleaning

Check for missing values, normalize column names, and clean the data.

```python
# Dataset shape and column inspection
print(df.shape)
print(df.columns)

# Normalize column names
df.columns = df.columns.str.strip().str.lower().str.replace('-', '_').str.replace(' ', '_')

# Replace '?' with NaN and drop missing values
df = df.replace('?', np.nan)
df = df.dropna()

df.info()
```

## 4. Basic Distributions

Understand the overall makeup of the dataset through simple counts and visualizations.

```python
# Income distribution
sns.countplot(data=df, x='income')
plt.title('Income Distribution')
plt.show()

# Gender distribution
sns.countplot(data=df, x='gender')
plt.title('Gender Distribution')
plt.show()
```

## 5. Key Questions & Analysis

Questions to understand how different features relate to income.

### Q1: Does Education Affect Income?

```python
# Show proportion of income levels per education group
edu_income = pd.crosstab(df['education'], df['income'], normalize='index') * 100
edu_income.sort_values(by='>50K', ascending=False).plot(kind='bar', stacked=True)
plt.title('Income by Education Level (%)')
plt.ylabel('Percentage')
plt.show()
```

### Q2: Is There a Gender Income Gap?

```python
# Compare income distribution by sex
gender_income = pd.crosstab(df['gender'], df['income'], normalize='index') * 100
gender_income.plot(kind='bar', stacked=True)
plt.title('Income by Gender (%)')
plt.ylabel('Percentage')
plt.show()
```

### Q3: Does Age Correlate with Income?

```python
# Show age distribution split by income
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='income', bins=30, kde=True, multiple='stack')
plt.title('Age Distribution by Income')
plt.show()
```

### Q4: Work Hours vs. Income?

```python
# Boxplot to compare working hours per income group
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='income', y='hours_per_week')
plt.title('Hours Worked per Week vs. Income')
plt.show()
```

## 6. Insights & Conclusion

We summarize the key insights derived from the analysis.

### Insights

- Education level strongly influences income; higher degrees lead to higher earnings.
- A gender gap exists: a greater share of men earn >$50K compared to women.
- Age correlates with income up to peak earning years (approx. 40–60).
- Higher work hours generally align with higher income, but not exclusively.

### Conclusion

Key socio-demographic factors such as education, gender, and age significantly influence income levels. These insights could inform workforce policy, training, and equality efforts.
